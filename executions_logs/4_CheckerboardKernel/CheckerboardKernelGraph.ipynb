```python
import re
import pandas as pd
import matplotlib.pyplot as plt

# Load the log with block_size parameter
file_path = "2025-06-12_15-40-32.log"
with open(file_path, "r") as file:
    log_data = file.read()

# Extract block_size and Best val acc
pattern = r"Filtered with filter Checkerboard Kernel with parameter: block_size = (\d+).*?Best val acc: (\d+\.\d+)%"
matches = re.findall(pattern, log_data, re.DOTALL)

# Create DataFrame
data_block = pd.DataFrame(matches, columns=["block_size", "best_val_acc"])
data_block = data_block.astype({"block_size": int, "best_val_acc": float})
data_block.sort_values("block_size", inplace=True)

# Plot 2D line chart
plt.figure(figsize=(10, 6))
plt.plot(data_block["block_size"], data_block["best_val_acc"], marker='o', linestyle='-', color='darkblue')
plt.title("Best Validation Accuracy vs Block Size")
plt.xlabel("Block Size")
plt.ylabel("Best Val Acc (%)")
plt.grid(True)
plt.tight_layout()
plt.show()
```
