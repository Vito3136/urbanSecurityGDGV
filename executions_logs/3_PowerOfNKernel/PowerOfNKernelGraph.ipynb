```python
import re
import pandas as pd
import matplotlib.pyplot as plt

# Load the new log file
file_path = "2025-06-11_12-18-30.log"
with open(file_path, "r") as file:
    log_data = file.read()

# Extract N and Best val acc
pattern = r"Filtered with filter Power of N Kernel with parameter: N = (\d+).*?Best val acc: (\d+\.\d+)%"
matches = re.findall(pattern, log_data, re.DOTALL)

# Create DataFrame
data_n = pd.DataFrame(matches, columns=["N", "best_val_acc"])
data_n = data_n.astype({"N": int, "best_val_acc": float})
data_n.sort_values("N", inplace=True)

# Plot 2D line chart
plt.figure(figsize=(10, 6))
plt.plot(data_n["N"], data_n["best_val_acc"], marker='o', linestyle='-', color='darkred')
plt.title("Best Validation Accuracy vs N Parameter")
plt.xlabel("N")
plt.ylabel("Best Val Acc (%)")
plt.grid(True)
plt.tight_layout()
plt.show()
```
