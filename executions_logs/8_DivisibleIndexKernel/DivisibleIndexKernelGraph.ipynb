```python
import re
import pandas as pd
import matplotlib.pyplot as plt


# Load the log file with single parameter: mod
file_path = "2025-06-12_14-11-24.log"
with open(file_path, "r") as file:
    log_data = file.read()

# Extract mod and Best val acc
pattern = r"Filtered with filter Divisible Index Kernel with parameter: mod = (\d+).*?Best val acc: (\d+\.\d+)%"
matches = re.findall(pattern, log_data, re.DOTALL)

# Create DataFrame
data_mod = pd.DataFrame(matches, columns=["mod", "best_val_acc"])
data_mod = data_mod.astype({"mod": int, "best_val_acc": float})
data_mod.sort_values("mod", inplace=True)

# Plot 2D line chart
plt.figure(figsize=(10, 6))
plt.plot(data_mod["mod"], data_mod["best_val_acc"], marker='o', linestyle='-', color='darkgreen')
plt.title("Best Validation Accuracy vs Mod Parameter")
plt.xlabel("Mod")
plt.ylabel("Best Val Acc (%)")
plt.grid(True)
plt.tight_layout()
plt.show()
```
