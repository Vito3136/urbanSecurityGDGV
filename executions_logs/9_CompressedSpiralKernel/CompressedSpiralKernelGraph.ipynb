```python
import re
import pandas as pd
import matplotlib.pyplot as plt

# Load the log file with the parameter "block"
file_path = "2025-06-12_10-31-35.log"
with open(file_path, "r") as file:
    log_data = file.read()

# Extract block and Best val acc
pattern = r"Filtered with filter Compressed Spiral Kernel with parameter: block = (\d+).*?Best val acc: (\d+\.\d+)%"
matches = re.findall(pattern, log_data, re.DOTALL)

# Create DataFrame
data_block = pd.DataFrame(matches, columns=["block", "best_val_acc"])
data_block = data_block.astype({"block": int, "best_val_acc": float})
data_block.sort_values("block", inplace=True)

# Plot 2D line chart
plt.figure(figsize=(10, 6))
plt.plot(data_block["block"], data_block["best_val_acc"], marker='o', linestyle='-', color='darkorange')
plt.title("Best Validation Accuracy vs Block Parameter")
plt.xlabel("Block")
plt.ylabel("Best Val Acc (%)")
plt.grid(True)
plt.tight_layout()
plt.show()
```
